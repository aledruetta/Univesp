```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graph(f, ini=0):
    n = 100
    plt.figure(figsize=(20, 5))
    plt.plot(range(ini, n), [f(a) for a in range(ini, n)], 'r.')
    plt.ylim(-1, 1)
    plt.show()
```

```python
def f(n):
    return 1 / n

def graph(f, k=0):
    n = 100
    plt.figure(figsize=(20, 5))
    plt.plot(range(k, n), [f(a) for a in range(k, n)], 'r.')
    plt.ylim(-1, 1)
    plt.show()
    
graph(f, 1)
```

```python
def f(n):
    return (n + 1) / (n**2 - 5)

def graph(f, k=0):
    n = 100
    plt.figure(figsize=(20, 5))
    plt.plot(range(k, n), [f(a) for a in range(k, n)], 'r.')
    plt.ylim(-1, 1)
    plt.show()
    
graph(f)
```

```python
def f(n):
    return (1 / n) - (1 / (n + 1))

def graph(f, k=0):
    n = 100
    plt.figure(figsize=(20, 5))
    plt.plot(range(k, n), [f(a) for a in range(k, n)], 'r.')
    plt.ylim(-1, 1)
    plt.show()
    
graph(f, k=1)
```

### Série Harmônica

```python
def f(n):
    return 1 / n

def graph1(f, k=0):
    n = 100
    plt.figure(figsize=(20, 5))
    plt.plot(range(k, n), [f(a) for a in range(k, n)], 'r.')
    plt.ylim(-1, 1)
    plt.show()

def graph2(f, k=0):
    n = 200
    plt.figure(figsize=(20, 5))
    plt.plot(range(k, n), [sum([f(j) for j in range(k, i)]) for i in range(k, n)], 'r.')
    plt.ylim(0, 6)
    plt.show()
    
graph1(f, k=1)
graph2(f, k=1)
```
