```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def createDF():
    df = pd.read_excel('Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)
    df.drop(['Type','Coverage','AREA','REG','DEV','DevName'], axis=1, inplace=True)
    df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
    df['Total'] = df[list(range(1980,2014))].sum(axis=1)
    df.set_index(['Country'], inplace=True)
    return df
```

```python
df_canada = createDF()
df_canada.head()
```

```python
def selectCountry(df, country):
    df.drop(['Continent','Region','Total'], axis=1, inplace=True)
    df = df.T
    return df[country]
```

```python
selectCountry(df_canada, 'Argentina').plot(kind='bar', figsize=(15,5))
plt.show()
```
